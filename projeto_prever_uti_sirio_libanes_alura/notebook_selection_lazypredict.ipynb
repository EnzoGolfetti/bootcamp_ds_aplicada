```python
!pip install lazypredict
from lazypredict.Supervised import LazyClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

###Lendo os datasets vindos do notebook principal

```python
data_corr = pd.read_csv("https://raw.githubusercontent.com/EnzoGolfetti/bootcamp_ds_aplicada/main/projeto_prever_uti_sirio_libanes_alura/dados_clean_corr.csv")
data_corr
```

```python
data_clean_binary_selected = pd.read_csv('https://raw.githubusercontent.com/EnzoGolfetti/bootcamp_ds_aplicada/main/projeto_prever_uti_sirio_libanes_alura/data_clean_binary_selected.csv')
data_clean_binary_selected
```

```python
data_clean_class_selected = pd.read_csv('https://raw.githubusercontent.com/EnzoGolfetti/bootcamp_ds_aplicada/main/projeto_prever_uti_sirio_libanes_alura/data_clean_class_selected.csv')
data_clean_class_selected
```

###Aplicando o LazyPredict

```python
#Lendo o dataset limpo de correlações
np.random.seed(8715)
data_corr = data_corr.sample(frac=1)
x = data_corr.drop(columns=['ICU'])
y = data_corr['ICU']

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.75, random_state=84552, stratify=y)

model_clf = LazyClassifier()
models,predictions = model_clf.fit(x_train, x_test, y_train, y_test)

models
```

```python
#Lendo o dataset limpo de correlações e variáveis binárias
np.random.seed(47582)
data_clean_binary_selected = data_clean_binary_selected.sample(frac=1)
x_b = data_clean_binary_selected.drop(columns=['ICU'])
y_b = data_clean_binary_selected['ICU']

x_train_b, x_test_b, y_train_b, y_test_b = train_test_split(x_b,y_b, train_size=0.75, random_state=781230, stratify=y)

model_clf = LazyClassifier()

models_b,predictions_b = model_clf.fit(x_train_b, x_test_b, y_train_b, y_test_b)

models_b
```

```python
#Lendo o dataset limpo de correlações e variáveis por classe
np.random.seed(89745653)
data_clean_class_selected = data_clean_class_selected.sample(frac=1)
x_c = data_clean_class_selected.drop(columns=['ICU'])
y_c = data_clean_class_selected['ICU']

x_train_c, x_test_c, y_train_c, y_test_c = train_test_split(x_c,y_c, train_size=0.75, random_state=781230, stratify=y)

model_clf = LazyClassifier()

models_c,predictions_c = model_clf.fit(x_train_c, x_test_c, y_train_c, y_test_c)

models_c
```

```python
pd.set_option('display.max_columns', None)
print('Performance por dataset dos três modelos com melhor desempenho',
      models.iloc[0:3,:-1],
      models_b.iloc[0:3,:-1],
      models_c.iloc[0:3,:-1], sep='\n')
```

Após rodarmos LazyPredict para o três modelos e utilizando o método sample do Pandas para diminuir a aleatoriedade juntamente com o train_test_spli method, os modelos selecionados para avançarmos foram:

- **RandomForestClassifier;**
- **LGBMClassifier;**
- **XGBClassifier;**
- **BernoulliNB.**

Observação: Decidimos selecionar 4 modelos e não apenas três pela diferença apresentadas entre os três datasets.
